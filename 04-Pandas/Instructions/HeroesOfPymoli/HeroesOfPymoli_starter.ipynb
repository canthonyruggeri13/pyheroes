
## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
grouped = purchase_data.groupby('Gender')

avgpriceGenderDF = pd.DataFrame()

##price column
avgpriceGenderDF["Average Purchase Price $"] = grouped["Price"].mean()

# purchase counts
avgpriceGenderDF["Purchase Count"] = grouped["Purchase ID"].count()


##purchase total per person 
avgpriceGenderDF["Total Purchase Value $"] = grouped["Price"].sum()

## avg spent per person
avgpriceGenderDF["Avg Total Purchase per Person $"] = grouped["Price"].sum() / grouped["SN"].nunique()

round(avgpriceGenderDF,2)
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
ageDF = purchase_data[['Age', 'Price', 'SN']]
bins = [0, 9, 14, 19, 24, 29, 34, 39, 500]
labels = ['<10', '10-14', '15-19', '20-24', ' 25-29',
        '30-34', '35-39', '40+']
ageDF['Age Bin'] = pd.cut(purchase_data['Age'], bins=bins, labels=labels)
```

```python
#Generate summary table from age_df

ageGroup = ageDF.groupby('Age Bin')
summary_df = pd.DataFrame()

# Count by Age Bin
summary_df['Count'] = ageGroup['Age'].count()

# Percentage in each Age Bin
summary_df['Pct'] = round( summary_df['Count'] / ageDF['Age'].count() * 100 , 2)
summary_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python

itemsDF = purchase_data[['Item ID', 'Item Name', 'Price']]
itemsDF.head()

grouped = itemsDF.groupby(['Item ID', 'Item Name'])

items_df = pd.DataFrame()


items_df['Purchase Count'] = grouped['Price'].count()


items_df['Item Price'] = grouped['Price'].mean()

#Total Purchase Value
items_df['Total Purchase Value'] = grouped['Price'].sum()

items_df.head()
```
